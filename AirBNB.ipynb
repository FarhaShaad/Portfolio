```python

# Remove duplicate data
df = df.drop_duplicates()
```

```python
# after dropping all the null value rows, log price have entry for 0.0 log_price. imputing with the median log_price of same property type
df = df[df['log_price']>0]
```

```python
# Standardising the cancellation_policy columns
df['cancellation_policy'] = df['cancellation_policy'].apply(lambda x: x.replace('super_strict_60', 'strict'))
df['cancellation_policy'] = df['cancellation_policy'].apply(lambda x: x.replace('super_strict_30', 'strict'))
```

```python
df.cancellation_policy.value_counts()
```

```python
# Standardising the host_has_profile_pic and host_identity_verified column to bool
df['host_has_profile_pic'] = df['host_has_profile_pic'].map({'t':True, 'f':False})
df['host_identity_verified'] = df['host_identity_verified'].map({'t':True, 'f':False})
df['instant_bookable'] = df['instant_bookable'].map({'t':True, 'f':False})
```

```python
# Dropping rows with zero number of beds
df.drop(index = df[df['beds'] == 0.0].index[0], axis=0, inplace=True)
```

```python
# Removing location related columns
df.drop(['name', 'neighbourhood', 'thumbnail_url', 'zipcode', 'id', 'description'], axis=1, inplace=True)
```

```python
#  Converting datatype of host_response_rate from string to int as its a percentage values
df['host_response_rate'] = df['host_response_rate'].apply(lambda x: int(x.split('%')[0]))
```

```python
df.host_response_rate.value_counts(normalize=True).head(8)
```

## 6.1 ML technique 1

1. Classification
    - Using Decision Tree Classification ML tecnique to predict the price range of the airbnb accommodation based on the features in the dataset.
    - we'll be splitting the dataset into train and test datasets to train the model and test the score of the prediction.

```python
dtree = DecisionTreeClassifier()
```

```python
# Splitting data into training and test data
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
# dividing training set further to get validation set to tune the decision tree
xtrain, xval, ytrain, yval = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
dtree.fit(xtrain, ytrain)
```

```python
ypred = dtree.predict(xtest)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(ytest, ypred)
cm = metrics.confusion_matrix(ytest, ypred)
accuracy_train = metrics.accuracy_score(ytrain, dtree.predict(xtrain))
print(f'training accuracy: {accuracy_train}')
print(f'test accuracy is {accuracy}')
print('confusion matrix:\n', cm)
```

Model is overfitting the data as training accuracy is 1 and test accuracy is .576

We can use pruning to avoid overfitting and make the model to generalize more

```python
accuracy_1 = []
for i in range(5,20):
    dtree_1 = DecisionTreeClassifier(max_depth = i)
    dtree_1.fit(xtrain, ytrain)
    accuracy_1.append(metrics.accuracy_score(yval, dtree_1.predict(xval)))
acc_df = pd.DataFrame({'max_depth':range(5,20), 'accuracy_1':accuracy_1})
sns.lineplot(data = acc_df, x='max_depth', y = 'accuracy_1')
```

- Validation Accuracy is rising till max_depth of 8 and decreases thereafter. So, we'll build the decision tree with max_depth as 8.

```python
dtree_1 = DecisionTreeClassifier(max_depth=8)
dtree_1.fit(xtrain, ytrain)

ypred_train = dtree_1.predict(xtrain)
accuracy_train = metrics.accuracy_score(ytrain, ypred_train)

ypred_val = dtree_1.predict(xval)
accuracy_val = metrics.accuracy_score(yval, ypred_val)

ypred_test = dtree_1.predict(xtest)
accuracy_test = metrics.accuracy_score(ytest, ypred_test)

f1_score = metrics.f1_score(y_true = ytest, y_pred = ypred_test, labels = dtree_1.classes_, average='macro')
precision = metrics.precision_score(y_true = ytest, y_pred = ypred_test, labels = dtree_1.classes_, average='macro', zero_division=np.nan)
recall = metrics.recall_score(y_true = ytest, y_pred = ypred_test, labels = dtree_1.classes_, average='macro')

print('Training accuracy score: {:.2f}\nValidation Accuracy score: {:.2f}\nTest Accuracy Score: {:.2f}'.format(accuracy_train, accuracy_val, accuracy_test))
print('\nPrecision score: {:.2f}\nRecall score: {:.2f}\nf1_score: {:.2f}'.format(precision, recall, f1_score))
```

## 6.2 ML technique 2 

2. Clustering
    - Using clutering, we can segment different airbnb accommodations and can be marketted that way.

```python
# Using kmeans clustering for clustering, and importing kmeans module for clustering
# Since, dataset have 19 features, we can reduce the dimensionality using PCA
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Reducing the dimensionality of the dataset using PCA
pca = PCA(n_components =2)
pca_nom = pca.fit_transform(X)
```

```python
# for deciding the optimal number of clustering, we can use the elbow curve
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, n_init='auto', max_iter=100)
    kmeans.fit(pca_nom)
    wcss.append(kmeans.inertia_)
wcss_df = pd.DataFrame({'cluster':list(range(1, 10)), 'wcss': wcss})
plt.figure(figsize=(6,6))
plt.plot(wcss_df['cluster'], wcss_df['wcss'], marker = 'o')
plt.show()
```

- we can infer from the elbow curve that it has inflection point at cluster number 3.
- So, we can cluster using 3 number of clusters.

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(pca_nom)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
cls = kmeans.predict(pca_nom)
```

```python
pca_nom[:,0].shape
```

```python
df_cluster = pd.DataFrame(pca_nom, columns = ['pca1', 'pca2'])
df_cluster['target'] = y
df_cluster['cluster'] = cls
sns.scatterplot(data=df_cluster, x='pca1', y='pca2', hue='cluster')
plt.show()
```
